```python
# Factorial Calculation Using Loop
num = int(input("Enter a number: "))
factorial = 1

if num < 0:
    print("Factorial does not exist for negative numbers")
else:
    for i in range(1, num + 1):
        factorial *= i
    print(f"Factorial of {num} is {factorial}")
```

```python
# Fibonacci Sequence Generation
n_terms = int(input("Enter the number of terms: "))
n1, n2 = 0, 1
count = 0

if n_terms <= 0:
    print("Please enter a positive integer")
elif n_terms == 1:
    print("Fibonacci sequence: 0")
else:
    print("Fibonacci sequence:")
    while count < n_terms:
        print(n1, end=" ")
        nth = n1 + n2
        n1 = n2
        n2 = nth
        count += 1
```

```python
# Check if a Number is Prime
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True

num = int(input("Enter a number: "))
if is_prime(num):
    print(f"{num} is a prime number")
else:
    print(f"{num} is not a prime number")
```

```python
# Frequency of Elements in List
lst = [1, 2, 2, 3, 4, 4, 4, 5, 5]
frequency = {}

for item in lst:
    frequency[item] = frequency.get(item, 0) + 1

print(frequency)
```

```python
# Find numbers divisible by 7 but not a multiple of 5
numbers = [str(num) for num in range(2000, 3201) if num % 7 == 0 and num % 5 != 0]
print(", ".join(numbers))
```

```python
# Count Upper and Lower Case Letters
sentence = input("Enter a sentence: ")
upper_case = 0
lower_case = 0

for char in sentence:
    if char.isupper():
        upper_case += 1
    elif char.islower():
        lower_case += 1

print(f"UPPER CASE {upper_case}")
print(f"LOWER CASE {lower_case}")
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

iris = pd.read_csv('/content/drive/MyDrive/Stats PGDM/Iris.csv')
iris
```

```python
iris.keys()
```

```python
iris.groupby('Species').agg({
    'SepalLengthCm': ['mean', 'std', 'max', 'median'],
    'SepalWidthCm': ['mean', 'std', 'max', 'median'],
    'PetalLengthCm': ['mean', 'std', 'max', 'median'],
    'PetalWidthCm': ['mean', 'std', 'max', 'median']
}).reset_index()
```

```python
# Group by species and calculate quartiles
iris.drop('Id', axis=1).groupby('Species').quantile([0.25, 0.5, 0.75])
```

```python
# Boxplots to compare distributions
plt.figure(figsize=(10, 6))
sns.boxplot(x='Species', y='SepalLengthCm', data=iris)
plt.title('Sepal Length by Species')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='Species', y='SepalWidthCm', data=iris)
plt.title('Sepal Width by Species')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='Species', y='PetalLengthCm', data=iris)
plt.title('Petal Length by Species')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='Species', y='PetalWidthCm', data=iris)
plt.title('Petal Width by Species')
plt.show()
```

```python
# Correlation matrix
correlation_matrix = iris.drop(['Id', 'Species'], axis=1).corr()
correlation_matrix
```

```python
# Heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.show()
```

```python
# Group by species and calculate correlation matrix for each species
correlation_matrix_grouped = iris.drop('Id', axis=1).groupby('Species').corr()

correlation_matrix_grouped
```

```python
for species, matrix in correlation_matrix_grouped.groupby('Species'):
  plt.figure(figsize=(8, 6))
  sns.heatmap(matrix, annot=True, cmap='viridis')
  plt.title(f"Feature Correlation Heatmap: {species}")
  plt.show()
```

```python
# Scatter plot: Petal Length vs Petal Width
plt.figure(figsize=(8, 6))
sns.scatterplot(x='PetalLengthCm', y='PetalWidthCm', hue='Species', data=iris, palette='deep')
plt.title('Petal Length vs Petal Width')
plt.show()

# Scatter plot: Sepal Length vs Sepal Width
plt.figure(figsize=(8, 6))
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', hue='Species', data=iris, palette='coolwarm')
plt.title('Sepal Length vs Sepal Width')
plt.show()
```
